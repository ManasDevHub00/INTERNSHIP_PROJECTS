```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('dataset/credit_risk_dataset.csv')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['person_emp_length'] = data['person_emp_length'].fillna(data['person_emp_length'].median())
```

```python
data['loan_int_rate'] = data['loan_int_rate'].fillna(data['loan_int_rate'].median())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
home_encoder = LabelEncoder()
intent_encoder = LabelEncoder()
grade_encoder = LabelEncoder()
default_encoder = LabelEncoder()
```

```python
data["person_home_ownership"] = home_encoder.fit_transform(data["person_home_ownership"])

data["loan_intent"] = intent_encoder.fit_transform(data["loan_intent"])

data["loan_grade"] = grade_encoder.fit_transform(data["loan_grade"])

data["cb_person_default_on_file"] = default_encoder.fit_transform(data["cb_person_default_on_file"])
```

```python
# data['person_home_ownership'] = encoder.fit_transform(data['person_home_ownership'])
# data['loan_intent'] = encoder.fit_transform(data['loan_intent'])
# data['loan_grade'] = encoder.fit_transform(data['loan_grade'])
# data['cb_person_default_on_file'] = encoder.fit_transform(data['cb_person_default_on_file'])

```

```python
data.head()
```

```python
x = data.drop('loan_status',axis=1)
y = data['loan_status']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scalar = StandardScaler()
```

```python
# X_train = scalar.fit_transform(X_train)
# X_test = scalar.fit_transform(X_test)

X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
print("accuracy")
print(accuracy_score(Y_test,model.predict(X_test)))
```

```python
from sklearn.metrics import  classification_report
```

```python
print("classification")
print(classification_report(Y_test,model.predict(X_test)))
```
